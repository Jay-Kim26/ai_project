## DATA 크기가 너무 많아 50만개로 줄임

```python
from sklearn.model_selection import train_test_split

n_samples = 500000

# stratify 옵션 사용해 비율 유지
df_small, _ = train_test_split(df,train_size=n_samples,stratify=df['Label'],random_state=42)

# 비율 확인 (정상적으로 잘 유지되는지 체크)
print(df['Label'].value_counts(normalize=True))
print(df_small['Label'].value_counts(normalize=True))
```

```python
cat_cols = ['proto', 'state', 'service','attack_cat']
label_encoders = {}
```

```python
from sklearn.preprocessing import LabelEncoder
for col in cat_cols:
    le = LabelEncoder()
    df_small[col + '_enc'] = le.fit_transform(df_small[col])
    label_encoders[col] = le
```

```python
exclude_cols = cat_cols + ['Label'] + [c + '_enc' for c in cat_cols] 
num_features = df_small.drop(columns=exclude_cols).columns.tolist() # 범주형 입력 변수 + 범주형 변수를 label encoding한 column + 타겟인 Label 제외
```

```python
df_small.reset_index(inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
scaler = MinMaxScaler()
X_num = scaler.fit_transform(df_small[num_features])
X_num = X_num.astype(np.float32) # 수치형 변수들의 값을 표준화
```

```python
X_proto = df_small['proto_enc'].astype(np.int64).values
X_state = df_small['state_enc'].astype(np.int64).values
X_service = df_small['service_enc'].astype(np.int64).values
y_attackcat = df_small['attack_cat_enc'].astype(np.int64).values
y_label = df_small['Label'].astype(np.int64).values # categorical 변수들은 모델에 따로 집어넣어야 하기 때문에 따로 저장, 타겟 변수 y_label
```

```python
X_num_train, X_num_test, \
X_proto_train, X_proto_test, \
X_state_train, X_state_test, \
X_service_train, X_service_test, \
y_label_train, y_label_test, \
y_attackcat_train, y_attackcat_test = train_test_split(
    X_num, X_proto, X_state, X_service,
    y_label, y_attackcat,
    test_size=0.2, 
    random_state=42, 
    stratify=y_label
)
# train / test set 분리, 8:2 비율, stratify =y_label로 타겟 변수 비율 맞춤
```

```python
import torch as torch
X_num_train_tensor = torch.tensor(X_num_train, dtype=torch.float32)
X_num_test_tensor = torch.tensor(X_num_test, dtype=torch.float32)

X_proto_train_tensor = torch.tensor(X_proto_train, dtype=torch.long)
X_proto_test_tensor = torch.tensor(X_proto_test, dtype=torch.long)

X_state_train_tensor = torch.tensor(X_state_train, dtype=torch.long)
X_state_test_tensor = torch.tensor(X_state_test, dtype=torch.long)

X_service_train_tensor = torch.tensor(X_service_train, dtype=torch.long)
X_service_test_tensor = torch.tensor(X_service_test, dtype=torch.long)

y_label_train_tensor = torch.tensor(y_label_train, dtype=torch.long)
y_label_test_tensor = torch.tensor(y_label_test, dtype=torch.long)

y_attackcat_train_tensor = torch.tensor(y_attackcat_train, dtype=torch.long)
y_attackcat_test_tensor = torch.tensor(y_attackcat_test, dtype=torch.long)
```

```python
from torch.utils.data import Dataset,DataLoader
class Dataset(Dataset):
    def __init__(self, X_num, proto, state, service, y_label,y_attackcat):
        self.X_num = X_num
        self.proto = proto
        self.state = state
        self.service = service
        self.y_label = y_label
        self.y_attackcat = y_attackcat
       

    def __len__(self):
        return len(self.y_attackcat)

    def __getitem__(self, idx):
        return (
            self.X_num[idx],
            self.proto[idx],
            self.state[idx],
            self.service[idx],
            self.y_label[idx],
            self.y_attackcat[idx]
        )


train_dataset = Dataset(
    X_num_train_tensor,
    X_proto_train_tensor,
    X_state_train_tensor,
    X_service_train_tensor,
    y_label_train_tensor,
    y_attackcat_test_tensor
)
test_dataset = Dataset(
    X_num_test_tensor,
    X_proto_test_tensor,
    X_state_test_tensor,
    X_service_test_tensor,
    y_label_test_tensor,
    y_attackcat_test_tensor
)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)


test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)

```

```python
import torch
import torch.nn as nn

class MLPWithEmbedding(nn.Module):
    def __init__(self, proto_classes, state_classes, service_classes):
        super().__init__()

        self.proto_emb = nn.Embedding(proto_classes, 4) 
        self.state_emb = nn.Embedding(state_classes, 4)
        self.service_emb = nn.Embedding(service_classes, 4)
 
        self.mlp = nn.Sequential(
            nn.Linear(36 + 12, 64),  # 36 numeric + 3*4 embedding
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, 2)  # Output: 2 classes (binary)
        )

    def forward(self, x_num, proto_idx, state_idx, service_idx):
        proto_vec = self.proto_emb(proto_idx)
        state_vec = self.state_emb(state_idx)
        service_vec = self.service_emb(service_idx)

        x = torch.cat([x_num, proto_vec, state_vec, service_vec], dim=1)
        return self.mlp(x)
```

```python
model = MLPWithEmbedding(
    proto_classes=len(label_encoders['proto'].classes_),
    state_classes=len(label_encoders['state'].classes_),
    service_classes=len(label_encoders['service'].classes_)
)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)

num_epochs = 5
model.train()

for epoch in range(num_epochs):
    
    for x_num, proto, state, service, label, _ in train_loader: # train_loader에는 6개 입력값이 있는데 attack은 필요 없으므로 _ 사용
        x_num = x_num.to(device)
        proto = proto.to(device)
        state = state.to(device)
        service = service.to(device)
        label = label.to(device).long()

        # Forward
        output = model(x_num, proto, state, service)
        loss = criterion(output, label)

        # Backward
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        

    print(f"Epoch {epoch+1}/{num_epochs} - Loss: {loss.item():.4f}")
```

```python
from sklearn.metrics import classification_report, accuracy_score

# 1단계 평가 + 2단계 데이터 구성
model.eval()
y_true_bin, y_pred_bin = [], []
x_num_list, proto_list, state_list, service_list, attackcat_list = [], [], [], [], []
normal_label_index = label_encoders['attack_cat'].transform(['Normal_packet'])[0] 
# 1단계로 attack인지 아닌지 판별하고 attack_cat 판별로 넘어가는데, attack_cat에는 Normal_packet이라는 column이 있으므로 이 데이터들을 2단계 입력값에서 제외해야함
with torch.no_grad():
    for x_num, proto, state, service, y_label, y_attackcat in test_loader:
        x_num = x_num.to(device)
        proto = proto.to(device)
        state = state.to(device)
        service = service.to(device)

        output = model(x_num, proto, state, service)
        y_pred = torch.argmax(output, dim=1)

        y_true_bin.extend(y_label.numpy())
        y_pred_bin.extend(y_pred.cpu().numpy())

        # 2단계 대상 추출 (1로 예측된 데이터)
        mask = (y_pred == 1) & (y_attackcat != normal_label_index)
        if mask.sum() > 0:
            x_num_list.append(x_num[mask].cpu())
            proto_list.append(proto[mask].cpu())
            state_list.append(state[mask].cpu())
            service_list.append(service[mask].cpu())
            attackcat_list.append(y_attackcat[mask].cpu())

print("=== [1단계 이진 분류 평가] ===")
print("Accuracy:", accuracy_score(y_true_bin, y_pred_bin))
print(classification_report(y_true_bin, y_pred_bin, digits=4))

```

```python
# 2단계용 데이터 통합
X_num_attack = torch.cat(x_num_list)
proto_attack = torch.cat(proto_list)
state_attack = torch.cat(state_list)
service_attack = torch.cat(service_list)
y_attackcat = torch.cat(attackcat_list)
```

## attackcat 데이터가 Generic이 압도적으로 많고, 소수 class의 비율이 너무 적기때문에 2차 multi-classification이 성능이 떨어질 수 있으므로 KMeasSMOTE 사용

```python
from imblearn.over_sampling import KMeansSMOTE

# 1. 텐서 → numpy 변환
X_num_np = X_num_attack.numpy()
proto_np = proto_attack.numpy()
state_np = state_attack.numpy()
service_np = service_attack.numpy()
y_attackcat_np = y_attackcat.numpy()

# 2. 범주형을 one-hot encoding
proto_df = pd.get_dummies(proto_np, prefix='proto')
state_df = pd.get_dummies(state_np, prefix='state')
service_df = pd.get_dummies(service_np, prefix='service')

X_cat_onehot = pd.concat([proto_df, state_df, service_df], axis=1).values

# 3. 수치형 + 범주형 결합
X_all = np.concatenate([X_num_np, X_cat_onehot], axis=1)
```

```python
# 4. KMeansSMOTE 적용
kmeans_smote = KMeansSMOTE(random_state=42, k_neighbors=5, cluster_balance_threshold=0.001)
X_resampled, y_resampled = kmeans_smote.fit_resample(X_all, y_attackcat_np)
```

```python
# 5. one-hot → 다시 인덱스로 복원
num_numerical = X_num_np.shape[1]
proto_end = proto_df.shape[1]
state_end = proto_end + state_df.shape[1]
service_end = state_end + service_df.shape[1]

proto_onehot = X_resampled[:, num_numerical:num_numerical + proto_end]
state_onehot = X_resampled[:, num_numerical + proto_end:num_numerical + state_end]
service_onehot = X_resampled[:, num_numerical + state_end:num_numerical + service_end]

proto_idx = np.argmax(proto_onehot, axis=1)
state_idx = np.argmax(state_onehot, axis=1)
service_idx = np.argmax(service_onehot, axis=1)
```

```python
# 6. 다시 텐서로 변환 (2단계 학습용)
X_num_attack_tensor = torch.tensor(X_resampled[:, :num_numerical], dtype=torch.float32)
proto_attack_tensor = torch.tensor(proto_idx, dtype=torch.long)
state_attack_tensor = torch.tensor(state_idx, dtype=torch.long)
service_attack_tensor = torch.tensor(service_idx, dtype=torch.long)
y_attack_attackcat_tensor = torch.tensor(y_resampled, dtype=torch.long)
```

```python
X_num_attack_train, X_num_attack_test, \
proto_attack_train, proto_attack_test, \
state_attack_train, state_attack_test, \
service_attack_train, service_attack_test, \
y_attack_attackcat_train, y_attack_attackcat_test = train_test_split(
    X_num_attack_tensor,
    proto_attack_tensor,
    state_attack_tensor,
    service_attack_tensor,
    y_attack_attackcat_tensor,
    test_size=0.2,
    random_state=42,
    stratify=y_attack_attackcat_tensor  # class 균형 유지하고 train test split
)
```

```python
y_label_attack_train = torch.ones_like(y_attack_attackcat_train)
y_label_attack_test = torch.ones_like(y_attack_attackcat_test)
```

```python
attack_train_dataset = Dataset(
    X_num_attack_train,
    proto_attack_train,
    state_attack_train,
    service_attack_train,
    y_label_attack_train,
    y_attack_attackcat_train
)

attack_test_dataset = Dataset(
    X_num_attack_test,
    proto_attack_test,
    state_attack_test,
    service_attack_test,
    y_label_attack_test,
    y_attack_attackcat_test
)

# 4. DataLoader 구성
attack_train_loader = DataLoader(attack_train_dataset, batch_size=128, shuffle=True)
attack_test_loader = DataLoader(attack_test_dataset, batch_size=128, shuffle=False)
```

```python
class MLPAttackClassifier(nn.Module):
    def __init__(self, proto_classes, state_classes, service_classes, num_attack_classes):
        super().__init__()
        self.proto_emb = nn.Embedding(proto_classes, 4)
        self.state_emb = nn.Embedding(state_classes, 4)
        self.service_emb = nn.Embedding(service_classes, 4)

        self.mlp = nn.Sequential(
            nn.Linear(36 + 12, 64),
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, num_attack_classes)
        )

    def forward(self, x_num, proto_idx, state_idx, service_idx):
        proto_vec = self.proto_emb(proto_idx)
        state_vec = self.state_emb(state_idx)
        service_vec = self.service_emb(service_idx)
        x = torch.cat([x_num, proto_vec, state_vec, service_vec], dim=1)
        return self.mlp(x)
```

```python
num_classes = len(label_encoders['attack_cat'].classes_)

model_attack = MLPAttackClassifier(
    proto_classes=len(label_encoders['proto'].classes_),
    state_classes=len(label_encoders['state'].classes_),
    service_classes=len(label_encoders['service'].classes_),
    num_attack_classes=num_classes
).to(device)

criterion_attack = nn.CrossEntropyLoss()
optimizer_attack = torch.optim.Adam(model_attack.parameters(), lr=0.001)

num_epochs = 5
model_attack.train()

for epoch in range(num_epochs):
    
    for x_num, proto, state, service, _, y_attackcat in attack_train_loader:
        x_num = x_num.to(device)
        proto = proto.to(device)
        state = state.to(device)
        service = service.to(device)
        y_attackcat = y_attackcat.to(device).long()

        output = model_attack(x_num, proto, state, service)
        loss = criterion_attack(output, y_attackcat)

        optimizer_attack.zero_grad()
        loss.backward()
        optimizer_attack.step()


    print(f"[2단계] Epoch {epoch+1}/{num_epochs} - Loss: {loss.item():.4f}")
```

```python
from sklearn.metrics import classification_report, accuracy_score

model_attack.eval()
y_true_attack, y_pred_attack = [], []

with torch.no_grad():
    for x_num, proto, state, service, _, y_attackcat in attack_test_loader:
        x_num = x_num.to(device)
        proto = proto.to(device)
        state = state.to(device)
        service = service.to(device)

        output = model_attack(x_num, proto, state, service)
        pred = torch.argmax(output, dim=1)

        y_true_attack.extend(y_attackcat.cpu().numpy())
        y_pred_attack.extend(pred.cpu().numpy())

# 출력

used_labels = np.unique(y_true_attack)
class_names_filtered = [label_encoders['attack_cat'].classes_[i] for i in used_labels] # Attack Cat이 Normal packet인 것은 제외했기 때문에
# 평가 결과 출력
print("=== [2단계 공격 유형 분류 평가] ===")
print(classification_report(
    y_true_attack,
    y_pred_attack,
    labels=used_labels,
    target_names=class_names_filtered,  # ← 사람이 읽을 수 있는 클래스 이름으로 변환
    digits=4
))

acc = accuracy_score(y_true_attack, y_pred_attack)
print(f"Accuracy: {acc:.4f}")
```
